```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
# Load an image using OpenCV (replace 'image.jpg' with your image file)
dark = cv2.imread('/content/drive/MyDrive/Image prossesing/histogram/dark.jpeg', cv2.IMREAD_GRAYSCALE)
white = cv2.imread('/content/drive/MyDrive/Image prossesing/histogram/white.jpeg', cv2.IMREAD_GRAYSCALE)
normal = cv2.imread('/content/drive/MyDrive/Image prossesing/histogram/normal.jpeg', cv2.IMREAD_GRAYSCALE)
```

```python
# Calculate the histogram using numpy
hist1 = cv2.calcHist([dark], [0], None, [256], [0, 256])
hist2 = cv2.calcHist([white], [0], None, [256], [0, 256])
hist3 = cv2.calcHist([normal], [0], None, [256], [0, 256])
```

```python

# Create a Matplotlib figure
plt.figure(figsize=(12, 6))

# Create subplots for each histogram
plt.subplot(1, 3, 1)
plt.plot(hist1)
plt.title('Dark Image Histogram')
plt.xlabel('Pixel Value')
plt.ylabel('Frequency')

plt.subplot(1, 3, 2)
plt.plot(hist2)
plt.title('White Image Histogram')
plt.xlabel('Pixel Value')
plt.ylabel('Frequency')

plt.subplot(1, 3, 3)
plt.plot(hist3)
plt.title('Normal Image Histogram')
plt.xlabel('Pixel Value')
plt.ylabel('Frequency')

# Adjust the layout and display the figure
plt.tight_layout()
plt.show()
```
