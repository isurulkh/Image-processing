```python
from PIL import Image, ImageDraw
import matplotlib.pyplot as plt

image = Image.open('/content/drive/MyDrive/Image prossesing/new.jpg')

plt.imshow(image)
plt.axis('off')
plt.show()
```

```python
# Reshape with resizing
resized_image = image.resize((1080, 960))

plt.imshow(resized_image)
plt.axis('off')
plt.show()
```

```python
# Spin and Turn with Rotation
rotated_image = resized_image.rotate(90, expand=True)


plt.imshow(rotated_image)
plt.axis('off')
plt.show()
```

```python
left = 100  # X-coordinate of the left edge of the region
upper = 50   # Y-coordinate of the upper edge of the region
right = 300  # X-coordinate of the right edge of the region
lower = 200  # Y-coordinate of the lower edge of the region
# Crop the image using the defined coordinates
cropped_image = rotated_image.crop((left, upper, right, lower))

plt.imshow(cropped_image)
plt.axis('off')
plt.show()
```

```python
# Visual Effects with Filters
from PIL import ImageFilter
filtered_image = cropped_image.filter(ImageFilter.GaussianBlur(radius=2))
plt.imshow(filtered_image)
plt.axis('off')
plt.show()
```

```python
blurred_image = filtered_image.filter(ImageFilter.BLUR)
plt.imshow(blurred_image)
plt.axis('off')
plt.show()
```

```python
sharpened_image = blurred_image.filter(ImageFilter.SHARPEN)

plt.imshow(sharpened_image)
plt.axis('off')
plt.show()
```

```python
# grayscale Color Space Exploration
grayscale_image = sharpened_image.convert('L')

plt.imshow(grayscale_image)
plt.axis('off')
plt.show()
```

```python
# Convert the image back to RGB
rgb_image = grayscale_image.convert('RGB')
print("RGB Image")

display(rgb_image)
```

```python
rgb_image.save('final_image.jpg')
```
