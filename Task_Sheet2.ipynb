```python
# Task 4: Removing Scratches and Blemishes
from PIL import ImageDraw

blemish_position = (100, 100)
blemish_radius = 5  # The size of the brush

# Use Pillow's drawing functions to edit the image and remove or conceal the imperfections.
draw = ImageDraw.Draw(image1)

display(image1)

```

```python
def restore_image(input_path, output_path):
    input_image = Image.open(input_path)
    blurred_image = input_image.filter(ImageFilter.GaussianBlur(radius=2))
    sharpened_image = blurred_image.filter(ImageFilter.SHARPEN)

    # Replace (100, 100) with the coordinates of the blemish to remove
    blemish_position = (100, 100)
    blemish_radius = 5
    draw = ImageDraw.Draw(sharpened_image)
    background_color = (255, 255, 255)
    draw.ellipse((blemish_position[0] - blemish_radius, blemish_position[1] - blemish_radius,
                  blemish_position[0] + blemish_radius, blemish_position[1] + blemish_radius), fill=background_color)

    sharpened_image.save(output_path)
```

```python
restore_image('/content/drive/MyDrive/Image prossesing/task2/image11.jpg', '/content/drive/MyDrive/Image prossesing/task2/restored_image.jpg')
```
